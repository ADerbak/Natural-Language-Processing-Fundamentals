```python
import urllib3
from pprint import pprint
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
http = urllib3.PoolManager()
rrq = http.request('GET', 'https://www.goodreads.com/search.xml', fields={"key":key,"q":"Bhagwad Gita"})
#q represents the querry we want to search for

rrq.status
```

```python
tree = ET.fromstring(rrq.data)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(rrq.data, 'lxml')
```

```python
pprint(soup)
```

```python
df = pd.DataFrame([(title.text.strip(),author.text.strip(),rating.text.strip())\
                   for title,author,rating in zip(soup.find_all('title'), \
                                                  soup.find_all('name'), soup.find_all('average_rating'))])
df.columns = ['Book Name', 'Author', 'Average Rating']
df
```
