# Sarcasm Detection

```python
import pandas as pd
```

```python
NEWS_DATASET = 'News-Headlines-Dataset-For-Sarcasm-Detection/Sarcasm_Headlines_Dataset.json'
```

```python
def parseJson(fname):
    for line in open(fname, 'r'):
        yield eval(line)
        
def get_data(fname):
    return pd.DataFrame(list(parseJson(NEWS_DATASET)))
```
